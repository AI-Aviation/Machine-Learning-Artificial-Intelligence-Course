# Metrics

## True Positives, False Negatives, False Positives, and True Negatives
Cat Classifier Example (A binary classifier)
- True Positive (TP)  - The classifier sees an image of a cat and classifies it as representing a cat
- False Negative (FN) - The classifier sees an image of a cat and classifies it as not representing a cat.
- False Positive (FP) - The classifier sees an image of something other than a cat but classifies it as representing a cat.
- True Negative (TN)  - The classifier sees an image of something other than a cat and classifies it as not representing a cat.

Some people will say 'predicts' or 'estimates' instead of 'classifies'.
For example, "the classifier predicted a cat" or "the classifier predicted it was not a cat"

For the binary classifier we want the False Positives and False Negatives to be zero.
That would be a perfect classifier.

Multi-class Classifiers are a little more complex

```python
display.Image(os.path.normpath("../Artwork/Confusion_matrix_KNearestNeighbors-MNIST-Fashion.jpg"))
```

A perfect classifier will have zeroes for all non-diagonal cell values.
We want the sum of the diagonal to equal the total number of samples.

```python
# Accuracy
# Accuracy is the percentage of all correctly classified observations.
# Accuracy can be misleading in large imbalanced data sets.

# For example, a three class multi-class classifier
# Dogs - 990 samples
# Cats - 5 samples
# Kangaroos - 5 samples

# Total number of samples: 1000

# If 900 observations of dogs are correct, and all the observations of cat sare correct,
# but none of the kangaroo observations were classified correctlly we would have an accuracy of:

# A = (900 + 5 + 0)/1000

(900 + 5 + 0)/1000
# 0.905 or 90.5%
```
